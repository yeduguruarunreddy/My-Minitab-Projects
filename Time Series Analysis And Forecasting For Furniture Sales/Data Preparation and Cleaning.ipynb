```python
df = pd.read_excel("Superstore.xls")
furniture = df.loc[df['Category'] == 'Furniture']
officesupplies = df.loc[df['Category'] == 'Office Supplies']
```

```python
cols = ['Row ID', 'Order ID', 'Ship Date', 'Ship Mode', 'Customer ID', 'Customer Name', 'Segment', 'Country', 'City', 'State', 'Postal Code', 'Region', 'Product ID', 'Category', 'Sub-Category', 'Product Name', 'Quantity', 'Discount', 'Profit']
furniture.drop(cols, axis=1, inplace=True)
furniture = furniture.sort_values('Order Date')
furniture.isnull().sum()
```

```python
furniture = furniture.groupby('Order Date')['Sales'].sum().reset_index()
furniture = furniture.set_index('Order Date')
furniture.index
```

```python
furniture_averages_daily_sales_value_for_month = furniture['Sales'].resample('MS').mean()
furniture_averages_daily_sales_value_for_month = furniture_averages_daily_sales_value_for_month.reset_index()
```

```python
# Current column names
current_columns = furniture_averages_daily_sales_value_for_month.columns

# Desired new column names
new_columns = ['Order Date','FurnitureAverageDailySalesForTheMonth']

furniture_averages_daily_sales_value_for_month.columns = new_columns
```

```python
# Extract Year and Month from the 'Order Date' column
furniture_averages_daily_sales_value_for_month['Year'] = furniture_averages_daily_sales_value_for_month['Order Date'].dt.year
furniture_averages_daily_sales_value_for_month['Month'] = furniture_averages_daily_sales_value_for_month['Order Date'].dt.month

# Convert month numbers to names
furniture_averages_daily_sales_value_for_month['Month'] = furniture_averages_daily_sales_value_for_month['Month'].apply(lambda x: calendar.month_abbr[x])

# Reorder the DataFrame if needed
furniture_averages_daily_sales_value_for_month = furniture_averages_daily_sales_value_for_month[['Year', 'Month', 'FurnitureAverageDailySalesForTheMonth']]
```

```python
officesupplies = officesupplies.groupby('Order Date')['Sales'].sum().reset_index()
```

```python
officesupplies = officesupplies.set_index('Order Date')
officesupplies.index
```

```python
officesupplies_averages_daily_sales_value_for_month = officesupplies['Sales'].resample('MS').mean()
officesupplies_averages_daily_sales_value_for_month = officesupplies_averages_daily_sales_value_for_month.reset_index()
```

```python
# Current column names
current_columns = officesupplies_averages_daily_sales_value_for_month.columns

# Desired new column names
new_columns = ['Order Date','OfficeSuppliesAverageDailySalesForTheMonth']

officesupplies_averages_daily_sales_value_for_month.columns = new_columns
```

```python
# Extract Year and Month from the 'Order Date' column
officesupplies_averages_daily_sales_value_for_month['Year'] = officesupplies_averages_daily_sales_value_for_month['Order Date'].dt.year
officesupplies_averages_daily_sales_value_for_month['Month'] = officesupplies_averages_daily_sales_value_for_month['Order Date'].dt.month

# Convert month numbers to names
officesupplies_averages_daily_sales_value_for_month['Month'] = officesupplies_averages_daily_sales_value_for_month['Month'].apply(lambda x: calendar.month_abbr[x])

# Reorder the DataFrame if needed
officesupplies_averages_daily_sales_value_for_month = officesupplies_averages_daily_sales_value_for_month[['Year', 'Month', 'OfficeSuppliesAverageDailySalesForTheMonth']]
```

```python
officesupplies_averages_daily_sales_value_for_month 
```

```python
salesdata = pd.merge(furniture_averages_daily_sales_value_for_month , officesupplies_averages_daily_sales_value_for_month , on=['Year', 'Month'], how='inner')
```
